```python
import numpy as np
import numpy.linalg as la
from sklearn import svm
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import cvxpy as cp
```

```python
X, y = make_blobs(n_samples=40, centers=2, random_state=6)

for i in range(len(y)):
    if y[i] == 0:
        y[i] = -1

# X = np.append(X, np.array([[8,-4],[5,-8]]), axis=0)
# y = np.append(y, np.array([1,0]))

# X = np.append(X, np.array([[7,-6]]), axis=0)
# y = np.append(y, np.array([1]))

X,y
```

```python
clf = svm.SVC(kernel='linear', C=10000)
# clf = svm.SVC(kernel='linear', C=0.2)
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
```

```python
print(clf.support_)
print('Margin: ', 2/la.norm(clf.coef_, 2))
for i in clf.support_:
    pred_val = clf.decision_function(X[i].reshape((1,-1)))
    pred_val2 = clf.coef_.dot(X[i]) + clf.intercept_
    
    print(pred_val, pred_val2, y[i])
    pred_label = clf.predict(X[i].reshape((1,-1)))
    if pred_label != y[i]:
        print('Misclassficiation')
        print(i, X[i], y[i], clf.decision_function(X[i].reshape((1,-1))), pred_label)
# print(clf.decision_function(np.array([[8,-4]])))
# print(clf.decision_function(np.array([[5,-8]])))
```

```python
r_0 = 0.3
r_1 = 0.1
X, y = make_blobs(n_samples=40, centers=2, random_state=6)

for i in range(len(y)):
    if y[i] == 0:
        y[i] = -1

clf = svm.SVC(kernel='linear', C=10000)
# clf = svm.SVC(kernel='linear', C=0.2)
clf.fit(X, y)

w = np.squeeze(clf.coef_)
b = clf.intercept_

plt.figure(figsize=(18, 16))
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

for i in range(X.shape[0]):
    if y[i] == -1:
        plt.plot(X[i,0]+r_0*np.cos(np.linspace(0,2*np.pi,100)),
            X[i,1]+r_0*np.sin(np.linspace(0,2*np.pi,100)), 'k')
    else:
        plt.plot(X[i,0]+r_1*np.cos(np.linspace(0,2*np.pi,100)),
            X[i,1]+r_1*np.sin(np.linspace(0,2*np.pi,100)), 'k')
        
ax = plt.gca()
xlim = np.array(ax.get_xlim())
ylim = ax.get_ylim()

y_val = (-w[0]*xlim-b)/w[1]
plt.plot(xlim, y_val)

ax.set_aspect('equal')

a_0 = cp.Variable(1)

cost = a_0
constr = []
for i in range(X.shape[0]):
    if y[i] == -1:
#         constr += [r_0*la.norm(w,2) + w.dot(X[i,:]) + b <= a_0]
        constr += [w.dot(X[i,:]) + b <= a_0]
        
problem = cp.Problem(cp.Minimize(cost), constr)
problem.solve(solver=cp.MOSEK, verbose=True)
a_0_val = -a_0.value
print(a_0_val)

y_0 = (-w[0]*xlim-(b-a_0.value))/w[1]

w_bar = -w
b_bar = -b

a_1 = cp.Variable(1)
cost = a_1
constr = []
for i in range(X.shape[0]):
    if y[i] == 1:
#         constr += [r_1*la.norm(w_bar,2) + w_bar.dot(X[i,:]) + b_bar <= a_1]
        constr += [w_bar.dot(X[i,:]) + b_bar <= a_1]
        
problem = cp.Problem(cp.Minimize(cost), constr)
problem.solve(solver=cp.MOSEK, verbose=True)

y_1 = (-w_bar[0]*xlim-(b_bar-a_1.value))/w_bar[1]

plt.plot(xlim, y_0)
plt.plot(xlim, y_1)
```
